```python

import requests
import numpy as np

def generate_embeddings(text):
    url = "https://api.euron.one/api/v1/euri/embeddings"
    headers = {
        "Content-Type": "application/json",
        "Authorization": "Bearer euri-f3a8bba1b99c7dff1d1fdddf69634c20b613941037ad64b039567262f062be7f"
    }
    payload = {
        "input": text,
        "model": "text-embedding-3-small"
    }

    response = requests.post(url, headers=headers, json=payload)
    data = response.json()

    embedding = np.array(data['data'][0]['embedding'])

    return embedding

```

```python
embeddings=[generate_embeddings(i) for i in data]
embeddings
```

```python
len(embeddings)
```

```python
collection=client.create_collection(name="viratkohli")
```

```python
collection.add(
    documents=data,
    embeddings=embeddings,
    ids=[str(i) for i in range(len(data))]
)
```

```python
collection.count()
```

```python
collection.get()
```

```python
query="who is virat kholi wife"
```

```python
query_embed= generate_embeddings(query).tolist()
query_embed
```

```python
collection.query(query_embeddings=[query_embed],n_results=2)
```
